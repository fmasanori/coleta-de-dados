```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'./MATRICULA_NORTE2014.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['ID_POSSUI_NEC_ESPECIAL'])
df2 = pd.read_csv(r'./MATRICULA_NORTE2015.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['IN_NECESSIDADE_ESPECIAL'])
df3 = pd.read_csv(r'./MATRICULA_NORTE2016.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['IN_NECESSIDADE_ESPECIAL'])
df4 = pd.read_csv(r'./MATRICULA_NORTE2017.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['IN_NECESSIDADE_ESPECIAL'])
df5 = pd.read_csv(r'./MATRICULA_NORTE2018.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['IN_NECESSIDADE_ESPECIAL'])
```

```python
Dados2014 = []
Dados2015 = []
Dados2016 = []
Dados2017 = []
Dados2018 = []
anos= ['2014', '2015', '2016', '2017', '2018']
Dados = [Dados2014, Dados2015, Dados2016, Dados2017, Dados2018]
```

```python
Dados2014.append(df1.query('(ID_POSSUI_NEC_ESPECIAL == 1)')['ID_POSSUI_NEC_ESPECIAL'].count())
Dados2015.append(df2.query('(IN_NECESSIDADE_ESPECIAL == 1)')['IN_NECESSIDADE_ESPECIAL'].count())
Dados2016.append(df3.query('(IN_NECESSIDADE_ESPECIAL == 1)')['IN_NECESSIDADE_ESPECIAL'].count())
Dados2017.append(df4.query('(IN_NECESSIDADE_ESPECIAL == 1)')['IN_NECESSIDADE_ESPECIAL'].count())
Dados2018.append(df5.query('(IN_NECESSIDADE_ESPECIAL == 1)')['IN_NECESSIDADE_ESPECIAL'].count())
```

```python
fig,ax = plt.subplots()
plt.rcParams['figure.figsize'] = (15,8)
plt.plot(anos, Dados, color='red')
plt.scatter(anos, Dados, color='red')
ax.plot(anos, Dados, label='Avanço')
ax.grid()
plt.legend()
plt.title('Quantidade de Crianças com necessidades especiais nas escola do Norte do Brasil')
plt.ylabel('NÚMERO DE ALUNOS')
plt.xlabel('ANOS')
plt.show()
```
